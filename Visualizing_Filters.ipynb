```python
model = load_model('/content/drive/MyDrive/DataSets/My_Model')  #loading our model
```

```python
model.summary()  #checking our model
```

```python
layers = model.layers  #list of layers used in our model
```

```python
#conv layers at 0,3,6,9
```

```python
conv_layers = [0,3,6,9]
```

```python
filters,biases = layers[0].get_weights() #Returns filters and weights for specified layers
```

```python
filters.shape  #The shape (3,3,3,32) indicates (filter_height,filter_width,no_of_channels,no_of_filters)
```

**[See this post for deeper understanding of how filters works in convolutional neural networks and how the shapes are assigned for filters](https://cs231n.github.io/convolutional-networks/)**

```python
biases.shape
```

```python
fig1 = plt.figure(figsize= (20,20))  #Setting the plot size
n_filters = filters.shape[3]
cols = 5
rows = math.ceil(n_filters/cols)+1
channel = 0 
for i in range(1,n_filters+1):
  f = filters[:,:,:,i-1]  #selecting the ith filter
  fig1 = plt.subplot(rows,cols,i)
  fig1.set_xticks([])
  fig1.set_yticks([])
  plt.imshow(f[:,:,channel],cmap = 'gray',interpolation = 'nearest') #plotting the filters belonging to specified channel
  plt.title(f"Filter {i}")    #Giving names for each filter
plt.suptitle('First Conv2D layer\'s 1st channel filters',size = 54)  #Setting the main title
plt.show()
```
